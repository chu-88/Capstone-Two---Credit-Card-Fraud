```python
# Summary statistics for Column1
summary_column1 = df['Amount'].describe()

# Summary statistics for Column2
summary_column2 = df['AverageIncomeExpendicture'].describe()

print("Summary statistics for Amount:")
print(summary_column1)
print()

print("Summary statistics for AverageIncomeExpendicture:")
print(summary_column2)
```

It looks like the amounts are listed in Nigerian Naira, we'll create two new columns, converting the currency to USD, for better contextual understanding of the values of each transaction.

```python
# Given exchange rate for 1 USD
exchange_rate = 0.0013

# Convert 'Amount' and 'AverageIncomeExpenditure' columns to USD
df['AmountUSD'] = df['Amount'] * exchange_rate
df['AverageIncomeExpendictureUSD'] = df['AverageIncomeExpendicture'] * exchange_rate
```

```python
# Summary statistics for Column1
summary_column1 = df['AmountUSD'].describe()

# Summary statistics for Column2
summary_column2 = df['AverageIncomeExpendictureUSD'].describe()

print("Summary statistics for Amount in USD:")
print(summary_column1)
print()

print("Summary statistics for AverageIncomeExpendicture in USD:")
print(summary_column2)
```

*Note: It is interesting that the mean and max of the Amounts charged are significantly higher than AverageIncomeExpendicture, this will be further investigated when looking at correlations

```python
# Categorizing the data into new columns, based on Amount ('Very Low', 'Low', 'Moderate', 'High', 'Extreme'), split into 5 bins

# Define the bin edges for 'CatAmount' and 'CatAverageIncomeExpenditure'
amount_bins = 10  # Number of bins
income_bins = 10  # Number of bins

# Define labels for each bin
amount_labels = ['Very Low', 'Low', 'Moderate', 'Below Average', 'Average',
                 'Above Average', 'High', 'Very High', 'Extremely High', 'Outlier']
income_labels = ['Very Low', 'Low', 'Moderate', 'Below Average', 'Average',
                 'Above Average', 'High', 'Very High', 'Extremely High', 'Outlier']

# Create 'CatAmount' and 'CatAverageIncomeExpenditure' columns with labels
df['CatAmount'] = pd.cut(df['AmountUSD'], bins=amount_bins, labels=amount_labels)
df['CatAverageIncomeExpendicture'] = pd.cut(df['AverageIncomeExpendictureUSD'], bins=income_bins, labels=income_labels)
```

```python
# Calculate the percentage of each category in 'CatAmount'
cat_amount_percentages = df['CatAmount'].value_counts(normalize=True) * 100

# Calculate the percentage of each category in 'CatAverageIncomeExpendicture'
cat_income_percentages = df['CatAverageIncomeExpendicture'].value_counts(normalize=True) * 100

# Print the percentages in order of the labeling for 'CatAmount'
print("Percentage distribution for 'CatAmount':")
for label in amount_labels:
    percentage = cat_amount_percentages[label]
    print(f"{label}: {percentage:.2f}%")

# Print the percentages in order of the labeling for 'CatAverageIncomeExpendicture'
print("\nPercentage distribution for 'CatAverageIncomeExpendicture':")
for label in income_labels:
    percentage = cat_income_percentages[label]
    print(f"{label}: {percentage:.2f}%")
```

## 4. Comparing Categorical Data against 'Outcome' Value

```python
# Print all column names of the DataFrame
print(df.columns)

# Print all column types of the DataFrame
print(df.dtypes)
```

```python
# Select numerical columns for analysis
columns_of_interest = ['CustomerAge', 'Amount',
       'AverageIncomeExpendicture', 'Outcome',
       'AmountUSD', 'AverageIncomeExpendictureUSD']

# Calculate the correlation matrix
correlation_matrix = df[columns_of_interest].corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```

I can't take too much away from this heatmap; other than that the transaction amount seems to be correlated relatively significantly with Outcomes.

Visualizing the distribution of Fraudulent outcomes by category:

```python
# List of categorical columns to visualize
categorical_columns = ['Marital Status', 'CardColour', 'CardType', 'Domain', 'Customer_City_Address']

# Loop through each categorical variable and create pie charts
for col in categorical_columns:
    # Group the data by the categorical column and Outcome, then calculate the percentage
    grouped = df.groupby([col, 'Outcome']).size().unstack()
    grouped = grouped.div(grouped.sum(axis=1), axis=0)
    
    # Create a pie chart for each category
    for index, row in grouped.iterrows():
        plt.figure(figsize=(6, 6))
        plt.pie(row, labels=row.index, autopct='%1.1f%%', startangle=140)
        plt.title(f'Distribution of Outcome by {col} ({index})')
        plt.show()
```

It indeed looks like Fraudulent charges are equally distributed amongst categories and may not be of much help.

How correlated are the NaN or now lableled 0, CustomerAge transactions to Fraudulent transactions?

```python
# Filter the DataFrame for transactions with CustomerAge of 0
filtered_df = df[df['CustomerAge'] == 0]

# Count the number of True (1) and False (0) outcomes
true_count = (filtered_df['Outcome'] == True).sum()
false_count = (filtered_df['Outcome'] == False).sum()

print("Number of True (1) outcomes:", true_count)
print("Number of False (0) outcomes:", false_count)

# Data for the pie chart
labels = ['True (1)', 'False (0)']
sizes = [true_count, false_count]

# Create a pie chart
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title("Outcome Distribution for CustomerAge 0")
plt.show()
```

It looks like the data matches and we'll want to keep the NaN/O CustomerAge data
